# [HW 1-1] Data Indexing
Extract sensor signal (without time column) from 0.02 seconds to 0.12 seconds by indexing the ExampleData in Github.

* Data link: 'https://github.com/purduelamm/purdue_me597_iiot/blob/main/ml_tutorial/ML1/ExampleData?raw=true'
* Index the data as DataFrame format
* Refer to ML1_Code1

```python
# Import the necessary packages
import pandas as pd
import numpy as np

# Load ExampleData from Github
path='https://github.com/purduelamm/purdue_me597_iiot/blob/main/ml_tutorial/ML1/ExampleData?raw=true'
ExampleData=pd.read_csv(path,sep=',',names=['time(s)', 'Acceleration(g)', 'Voltage(V)', 'Current(kA)'])
ExampleData


# Data indexing
SensorDataOnly=ExampleData.iloc[:,1:]

StartPoint=np.where(ExampleData.iloc[:,0].values==0.02)[0][0]
EndPoint=np.where(ExampleData.iloc[:,0].values==0.12)[0][0]
# Define the indexed data as 'SensorData'
SensorData = SensorDataOnly.iloc[StartPoint:EndPoint ,:]

SensorData
```

# [HW 1-2] Data Processing & Plotting
Plot the change in mean value every 0.01 seconds for each sensor signal in 'SensorData' above.
* Refer to ML1_Code3

```python
# Import (additional) packages

import matplotlib.pyplot as plt

# Feature extraction
WindowSize=128 # Samplingfrequency is 12800
Mean_Values=np.zeros((int(SensorData.shape[0]/WindowSize),SensorData.shape[1]))
for  sensor in range(SensorData.shape[1]):
  for time in range(Mean_Values.shape[0]):
    Mean_Values[time,sensor] = np.mean(SensorData.iloc[WindowSize*time:WindowSize*(time+1),sensor])
TimeArr=np.arange(3,13) / 100

# Plot graphs
plt.figure(figsize=(10,5))
plt.plot(TimeArr,Mean_Values[:,0],ls='--',c='r',marker='o',mfc='r',mec='r') # Acceleration
plt.plot(TimeArr,Mean_Values[:,1],ls='--',c='g',marker='o',mfc='g',mec='g')# Voltage
plt.plot(TimeArr,Mean_Values[:,2],ls='--',c='b',marker='o',mfc='b',mec='b') # Current
plt.grid()
plt.xlabel('time(s)')
plt.ylabel('Mean value')
plt.legend(['Acceleration(g)', 'Voltage(V)', 'Current(kA)'],loc='best',fontsize=10)
plt.show()
```
