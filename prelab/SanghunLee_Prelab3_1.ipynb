Seamless real-time communication is the core prerequisite for all Industry 4.0 and Industrial Internet of
Things (IIoT) implementations. In Lab3, we will collect data from two IoT devices, a power meter ([Acuvim L series](https://www.accuenergy.com/products/acuvim-l-multifunction-power-energy-meter/), AccuEnergy, Canada), and a vibration sensor ([VVB001](https://www.ifm.com/us/en/product/VVB001), IFM, Germany) with a IO-Link master ([AL1302](https://www.ifm.com/us/en/product/AL1302), IFM, Germany). Please visit and look into both web pages of the devices. For the power meter, Modbus protocol will be used. On the other hand, for the vibration sensor, the IO-Link protocol will be used. When we read data from the vibration sensor through the IO-Link master, we use REST API via TCP/IP (so-called EtherNet/IP) communication. But the communication between the vibration sensor and the IO-Link master is being made by IO-Link protocol. This prelab is to install a related package on Raspberry Pi and to be familiar with the industry standard protocols. Modbus and IO-Link are representative protocols in Fieldbus communications. #### 4) Explain what is REST API?



---

A REST API is a method for applications to communicate over HTTP. It allows devices to request, send, update, or delete data using methods such as GET, POST, PUT, and DELETE.

#### 5) What can be the applications of power and vibration monitoring in the industry?



---
