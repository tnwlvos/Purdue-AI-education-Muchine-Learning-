## 1.3 Raspberry Pi Network Stability

In this section we will enhance the local network connectivity between the Raspberry Pi and your local PC using SSH. Follow the steps below.

0. From 1.2, you would have already accessed the CLI (Command Line Interface) of your Raspberry Pi in your PC Windows terminal.
1. Type the following in the CLI. This will allow you to change your Raspberry Pi's static ip and get restore default settings.

    `sudo nmcli con modify "netplan-eth0" ipv4.method manual ipv4.addresses 192.168.50.2/24 ipv4.gateway "" ipv4.dns "" ipv4.never-default yes`

2. Type the following in CLI. This will restart the network connection.

    `sudo nmcli con down "netplan-eth0"; sudo nmcli con up "netplan-eth0"`

3. Now in your PC, change your PC's ip (networked with Raspi) as described in the figure below.

<img src="https://github.com/purduelamm/purdue_me597_iiot/blob/main/lab/img_S26_summer/Lab0_fig3.png?raw=true" width="100%">

*Figure 2 PC ethernet ip settings*

4. Open a new termial (cmd) in your PC and run below. This will allow pinging from raspi.

    `netsh advfirewall firewall add rule name="Allow Ping from Raspberry Pi" dir=in action=allow protocol=icmpv4:8,any remoteip=192.168.50.2 profile=any`

## 1.6 Internet Connection with VNC

We will now connect to the internet in VNC. Follow the steps below.

**[Connection to PAL3.0 - for on-campus use]

0.  Notice that there is a Up & Down arrow in the top-right corner of your VNC screen, this is the internet settings. Check if have "PAL3.0"

1.  run the code below. This is delete all initial settings related to internet connection using "PAL3.0"

    `sudo nmcli con delete "PAL3.0" 2>/dev/null;`

2.  run the code below. make sure to change "<your_purdue_ID>" in the text.

    `sudo nmcli con add type wifi ifname wlan0 con-name "PAL3.0" ssid "PAL3.0" wifi-sec.key-mgmt wpa-eap 802-1x.eap peap 802-1x.identity “<your_purdue_ID>" 802-1x.phase2-auth mschapv2 802-1x.system-ca-certs no && sudo nmcli --ask con up "PAL3.0"`

3.  Change the wifi settings as noted in the figure below.

## 1.8 Looking Forward to Lab 0.2: Network-Based Remote Access

In Lab 0.1, the laptop and Raspberry Pi were connected primarily through a direct Ethernet connection. Lab 0.2 introduces two network tools used for different purposes:

- **Purdue Cisco VPN:** Allows an off-campus or online student to enter the Purdue network logically and access resources or devices that are available within the Purdue network.
- **Tailscale VPN:** Creates a private device-to-device network between the laptop and Raspberry Pi. It allows the devices to communicate through their Tailscale addresses even when direct communication is restricted on a secured network such as PAL 3.0.
